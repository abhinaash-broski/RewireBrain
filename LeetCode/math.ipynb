1. Sort the array in any order 
2. Check the diffference between them. 
3. If Same and equal. return True else False. 

```python
from typing import List

class Solution:
    def canMakeArithmeticProgression(self, arr: List[int]) -> bool:
        arr.sort()
        diff = arr[1] - arr[0]

        for i in range(2, len(arr)):
            if arr[i] - arr[i - 1] != diff:
                return False
        return True
```

```python
class Solution:
    def pivotInteger(self, n: int) -> int:
        total_sum = n * (n + 1) // 2
        left_sum = 0

        for i in range(1, n + 1):
            left_sum += i
            if left_sum == total_sum - left_sum + i:
                return i
        return -1
```
